```python
from pybit import HTTP
import pprint

with open("bybit.key") as f:
    lines = f.readlines()
    api_key = lines[0].strip()
    api_secret = lines[1].strip()

session = HTTP(
    endpoint="https://api.bybit.com", 
    api_key=api_key, 
    api_secret=api_secret
)

balances = session.get_wallet_balance()
pprint.pprint(balances)
```

```python
dt = datetime.datetime(2022, 7, 1, 0, 0, 0)
from_time = time.mktime(dt.timetuple())
for_time = 20

for i in range(for_time):
    resp = session.query_kline(
        symbol="BTCUSDT",
        interval=1,
        limit=200,
        from_time=from_time+i*60*200)

    if i == 0 :
        result = resp['result']
        dataset = pd.DataFrame(result)
        ts = pd.to_datetime(dataset['open_time'], unit='s')
        dataset.set_index(ts, inplace=True)
        dataset = dataset[['open', 'high', 'close','low','volume']]
    
    else:
        result = resp['result']
        df = pd.DataFrame(result)
        ts = pd.to_datetime(df['open_time'], unit='s')
        df.set_index(ts, inplace=True)
        df = df[['open', 'high', 'close','low','volume']]
        dataset = pd.concat([dataset,df])
```

```python
dataset.to_csv("temp.csv")
```
